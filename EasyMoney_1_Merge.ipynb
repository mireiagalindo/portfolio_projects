#### Shift PROD columns

Our goal is to predict whether a customer will buy a product or not. In order to do that, we need to work with past data. That is why we are shifting the PROD data for each client. We want to observe the client's profile before they make the purchase.

```python
# Dataframe for a random customer
df_customer_products[df_customer_products["pk_cid"] == 	1375586].T
```

```python
# Shift PROD columns
for col in PROD_col:
  df_customer_products[f"{col}_lag1"] = df_customer_products.groupby("pk_cid")[col].shift(1)
```

```python
# Verify shift
df_customer_products[df_customer_products["pk_cid"] == 	1375586].T
# After the shift, the first month that a client appears in our dataframe, all PROD-columns are NaN. 
# It is OK to replace this null values with "cero". 
```

```python
df_customer_products.isnull().sum()
```

```python
# Delete original columns
df_customer_products.drop(PROD_col, axis = 1, inplace = True)
```

```python
# Replace null values in PROD_lag1 columns with 0. 
PROD_col = [col for col in df_customer_products.columns if col.startswith('PROD')]
PROD_col

df_customer_products[PROD_col] = df_customer_products[PROD_col].fillna(0)
```

### Preliminary Feature Engineering

```python
df1 = df.copy()
```

```python
# Convert date columns
df1['SALE_month_sale'] = pd.to_datetime(df1['SALE_month_sale'], format='%Y-%m-%d')

df1["pk_partition"] = pd.to_datetime(df1['pk_partition'], format='%Y-%m')

df1["CUST_entry_date"] = pd.to_datetime(df1['CUST_entry_date'], format='%Y-%m')
```

```python
df1.tail().T
```

```python
# Create boolean indicator "new cliente" ( "nuevo_cliente" )
df1["nuevo_cliente"] = np.where(df1["pk_partition"] == df1["CUST_entry_date"], 1, 0)
```

```python
#create new indicator: client tenure in months ("antiguedad_cliente_meses" )
df1["antiguedad_cliente_dias"] = (df1["pk_partition"] - df1["CUST_entry_date"])
df1['antiguedad_cliente_dias'] = df1['antiguedad_cliente_dias'].dt.days
```

```python
df1['antiguedad_cliente_meses']= round(df1['antiguedad_cliente_dias']/30).astype(int)
df1.drop("antiguedad_cliente_dias", axis = 1, inplace = True)
```

#### TARGET

```python
# Create a new boolean indicator "PURCHASE" (indicating whether the client purchased any product in a particular month or not). ("HAY_VENTA") 
# That indicator is our TARGET, which we will aim to predict when we run the model.
df1["HAY_VENTA"] = np.where((df1["SALE_pk_sale"].notnull()), 1, 0)
```

```python
# For date columns, keep only the date, not the time
df1['SALE_month_sale'] = df1['SALE_month_sale'].dt.date

df1["pk_partition"] = df1["pk_partition"].dt.date

df1["CUST_entry_date"] = df1["CUST_entry_date"].dt.date
```
